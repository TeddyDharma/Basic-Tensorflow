```python
import gym
```

```python
#  membbuat frozen environtment
env = gym.make('FrozenLake-v1')
#  'FrozenLake-v0' sudah deprciated dan sudah digantikan dengan 'FrozenLake-v1'
#  dokumentasi untuk gym : https://www.gymlibrary.dev/
```

```python
# mendapatkan ada berapa  states di dalam model env
print(env.observation_space.n)
# mendapatkan berapa ada action : 
print(env.action_space.n)
```

```python
#  melakukan reset
env.reset()
```

```python
action = env.action_space.sample()
#  random action 
print(action)
```

```python
bservation, reward, terminated, truncated, info = env.step(action)
# di dokumentasi harus ada 5 variabel yang akan menampung kembalian 
```

```python
env.render()
#  Fungsi render() dalam Gym memiliki peran dalam menampilkan representasi visual dari lingkungan atau tugas yang sedang dijalankan oleh agen pembelajaran penguatan.
```
